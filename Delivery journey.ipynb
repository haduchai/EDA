```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# inore the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# read data
df = pd.read_csv('../data_remake\Delivery journey.csv')
df.head()
```

```python
# vẽ biểu đồ tròn thể hiện hiệu suất giao hàng của tất cả gian h
df['Order Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(10, 10))
plt.title('Order Status')
plt.ylabel('')
plt.show()
```

```python
# vẽ biểu đồ tròn thể hiện hiệu suất của tất cả các gian hàng
for store in df['store_name'].unique():
    df[df['store_name'] == store]['Order Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(10, 10))
    plt.title(store, fontsize=20)
    plt.ylabel('')
    plt.show()
```

```python
# hiệu suất bán hàng ở 5 địa điểm có tần suất mua hàng nhiều nhất
for location in df['Destination_Location'].value_counts().index[:5]:
    df[df['Destination_Location'] == location]['Order Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(10, 10))
    plt.title(location, fontsize=20)
    plt.ylabel('')
    plt.show()
```
